````python
products = ["t-shirt","mug","hat","book","keychain"]
````

````python
def initialize_inventory(products):
    inventory = {}
    for key in products:  
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f'stock of {key}: '))
                if quantity >= 0:
                    inventory[key] = quantity
                    valid_quantity = True
                else:
                    print('Quantity can not be negative. Please enter a valid quantity')
            except ValueError:
                print('Invalid input. Please enter a valid quantity')
    return inventory
````

````python
def get_customer_orders():

    customers_orders = set()

    def add_products(customers_orders, products):
        valid_product = False
        while not valid_product:
            try:
                producto = input("Añade un producto a tu pedido: ")
                if producto in products:
                    customers_orders.add(producto)
                    valid_product = True
                else:
                    print('No tenemos este producto. Añade un producto que este en la siguiente lista: t-shirt, mug,  hat, book, keychain ')
            except Exception as e:
                print(f'Error: {e}. Intenta de nuevo.')
                
    add_products(customers_orders, products)
    decision = input('¿Quieres añadir otro producto?: ').lower()

    while decision not in ('si', 'no'):
        try:
            decision = input('Por favor, escriba si o no: ').lower()
        except Exception as e:
                print(f'Error: {e}. Tienes que indicar si o no.')

    while decision == 'si':
        add_products(customers_orders, products)
        decision = input('¿Quieres añadir otro producto?: ').lower()
    
        while decision not in ('si', 'no'):
            try:
                decision = input('Por favor, escriba si o no: ').lower()
            except Exception as e:
                print(f'Error: {e}. Tienes que indicar si o no.')
    return customers_orders
````

````python
from functools import reduce

def addition(a, b):
    return a + b

def price_of_products(customers_orders):
    price_cust_ord = []

    for product in customers_orders:
        while True: #Bucle para seguir pidiendo el precio
            try:
                price = int(input(f"price of {product}: "))
                if price <= 0:
                    raise ValueError("El valor tiene que ser mayor a cero")
                price_cust_ord.append(price)
                break
                
            except ValueError as e:
                print(f'Error: {e}. Tienes que introducir un valor numérico.')
    
    total_price = reduce(addition, price_cust_ord, 0)
    return total_price
````

````python
def update_inventory(inventory, customers_orders):
    for product in customers_orders:
        if product not in inventory:
            print(f'El producto {product} no está en el inventario.')
            continue
        
        if inventory[product] == 0:
            print(f'No hay más existencias de {product}.')
            continue
        
        inventory[product] -= 1
    

    inventory = {key: value for key, value in inventory.items() if value > 0}
    return inventory
````

````python
def calculate_order_statistics(customers_orders, products):

    total_products_ordered=len(customers_orders)
    total_products = len(products)

    #Para evitar dividir por cero

    if total_products == 0:
        percentage_ordered = 0.0
    else:
        percentage_ordered = total_products_ordered / total_products * 100

    return percentage_ordered, total_products_ordered
````

````python
def print_order_statistics(percentage_ordered, total_products_ordered):
    
    print(f'El porcentaje de productos ordenados es: {percentage_ordered}')

    print(f'Order statistics \nTotal products ordered: {total_products_ordered} \nPercentage of products ordered: {percentage_ordered} %')
````

````python
inventory = initialize_inventory(products)

customers_orders = get_customer_orders()

update_inventory(inventory, customers_orders)

calculate_order_statistics(customers_orders, products)

percentage_ordered, total_products_ordered = calculate_order_statistics(customers_orders, products)

print_order_statistics(percentage_ordered, total_products_ordered)

print('Update inventory', *inventory.items(), sep='\n')

total_price = price_of_products(customers_orders)
print("Total price:", total_price)
````
